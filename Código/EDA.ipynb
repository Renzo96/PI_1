```python
#En consecuencia elimnamos la columna country y date_added por tener gran cantidad de faltantes. La columna 
# director no ya que puede ser una variable sensible para el estudio

plataformas.drop(columns = ['country', 'date_added'], inplace=True)

plataformas
```

```python
#Voy a colocar ND(No director) donde tenga los valores faltantes

plataformas.director.replace(np.nan,'ND', inplace = True)

#Verifico

plataformas.head()
```

```python
#Visualizo sus valores
plataformas.duration_type.value_counts()
```

```python
#importo libreria y el modulo que voy a usar
from sklearn.impute import SimpleImputer
```

```python
#Defino la estrategia
imputer = SimpleImputer(strategy='constant', fill_value=0)
```

```python
#Entreno
imputer.fit(plataformas[['duration_int']])

plataformas['duration_int'] = imputer.transform(plataformas[['duration_int']])
```

```python
#Visualizo
plataformas.isnull().sum()
```

```python
#Vamos a colocar NC donde cast tenga valores faltantes
plataformas.cast.replace(np.nan, 'NC', inplace = True)
```

```python
#Hago lo mismo que hice con cast pero reemplazando con min
plataformas.fillna(value='min', inplace=True)
```

```python
#Corroboramos
plataformas.isnull().sum()
```
