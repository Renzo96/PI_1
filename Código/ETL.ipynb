```python
# Importamos las librerias pertinentes

import pandas as pd
import numpy as np
```

```python
# Generamos las requests 

Amazon = pd.read_csv('../Datasets/plataformas/amazon_prime_titles.csv')
Disney = pd.read_csv('../Datasets/plataformas/disney_plus_titles.csv')
Hulu = pd.read_csv('../Datasets/plataformas/hulu_titles.csv')
Netflix = pd.read_csv('../Datasets/plataformas/netflix_titles.csv')
```

```python
#Creamos una lista de las plataformas de stream

lista_1 = [Amazon, Disney, Hulu, Netflix]
```

```python
#Eliminamos espacios vacíos

for df in lista_1:
    df['date_added'] = df['date_added'].str.strip()

# Netflix['date_added'] = Netflix['date_added'].str.strip()
```

```python
#Reemplazamos los nulos por 'G'

for df in lista_1:
    df['rating'].fillna('G', inplace=True)
```

```python
# Creamos una columna id
    
for df in lista_1:
    df.insert(0, 'id', '')
    
```

```python
#Corroboramos

Amazon.head()
```

```python
#Creamos una función que agregará la inicial a cada plataforma a la columna id

def add_a(x):
    return 'a' + str(x)

def add_d(x):
    return 'd' + str(x)

def add_h(x):
    return 'h' + str(x)

def add_n(x):  
    return 'n' + str(x)
```

```python
# Agregamos la primera letra de cada streaming al campo id


add_functions = [add_a, add_d, add_h, add_n]

for i, df in enumerate(lista_1):
    
    df['id'] = df['id'].apply(add_functions[i])
    df['id'] = df.apply(lambda row: row['id'] + str(row['show_id']), axis=1)
```

```python
#Corroboramos

Disney.head()
```

```python
#Cambiamos el campo fecha al formato especificado

for df in lista_1:
    df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y')
    df['date_added'] = df['date_added'].dt.strftime('%Y-%m-%d')
    df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
#Pasamos todo a minúscula

Amazon = Amazon.applymap(lambda x: x.lower() if type(x) == str else x)
Disney = Disney.applymap(lambda x: x.lower() if type(x) == str else x)
Hulu = Hulu.applymap(lambda x: x.lower() if type(x) == str else x)
Netflix = Netflix.applymap(lambda x: x.lower() if type(x) == str else x)
```

```python
#Creamos una función para pasar los float NaN a Integer

def split_duration(duration):
    if pd.isna(duration):
        return duration, duration
    value, unit = duration.split(" ")
    return int(value), unit
```

```python
#Cambio el campo duration en duration_int y duration_type( El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

lista_2 = [Amazon, Disney]

for df in lista_2:
    df[['duration_int', 'duration_type']] = df['duration'].str.split(expand=True)
    df['duration_int'] = df['duration_int'].astype(int)

    

```

```python
#Hago lo mismo pero con Hulu

Hulu[['duration_int', 'duration_type']] = Hulu['duration'].apply(split_duration).apply(pd.Series)

# Reemplaza los valores nulos con 0
Hulu['duration_int'] = Hulu['duration_int'].fillna(0)

# Redondea los valores a un entero
Hulu['duration_int'] = Hulu['duration_int'].apply(lambda x: int(round(x)))

# Cambia el tipo de datos a int
Hulu['duration_int'] = Hulu['duration_int'].astype(int)

# Hulu = Hulu.drop(columns='duration')

#vuelvo a reemplazar los valores 0 por Nan

Hulu['duration_int'] = np.where(Hulu['duration_int'] == 0, np.nan, Hulu['duration_int'])
```

```python
#Ahora con Netflix

Netflix[['duration_int', 'duration_type']] = Netflix['duration'].apply(split_duration).apply(pd.Series)

# Reemplaza los valores nulos con 0
Netflix['duration_int'] = Netflix['duration_int'].fillna(0)

# Redondea los valores a un entero
Netflix['duration_int'] = Netflix['duration_int'].apply(lambda x: int(round(x)))

# Cambia el tipo de datos a int
Netflix['duration_int'] = Netflix['duration_int'].astype(int)
```

```python
# Dropeamos la columna duration

Amazon = Amazon.drop(columns='duration')
Disney = Disney.drop(columns='duration')
Hulu = Hulu.drop(columns='duration')
Netflix = Netflix.drop(columns='duration')

```

```python
#Añado una columna platform a todos los datasets

# create the list of dataframes
lista_1 = [Amazon, Disney, Hulu, Netflix]

# create a list of platform names
platform_names = ['amazon', 'disney', 'hulu', 'netflix']

# use a for loop to add a "platform" column to each dataframe
for i in range(len(lista_1)):
    lista_1[i]['platform'] = platform_names[i]

```

```python
#Concateno los datasets en uno

peliculas = pd.concat([Amazon,Disney,Hulu,Netflix], axis=0)
```

```python
# Ahora procedemos con la carpeta Ratings

ratings1 = pd.read_csv('../Datasets/ratings/1.csv')
ratings2 = pd.read_csv('../Datasets/ratings/2.csv')
ratings3 = pd.read_csv('../Datasets/ratings/3.csv')
ratings4 = pd.read_csv('../Datasets/ratings/4.csv')
ratings5 = pd.read_csv('../Datasets/ratings/5.csv')
ratings6 = pd.read_csv('../Datasets/ratings/6.csv')
ratings7 = pd.read_csv('../Datasets/ratings/7.csv')
ratings8 = pd.read_csv('../Datasets/ratings/8.csv')

```

```python
#Concatenamos y chequemaos el shape

ratings = pd.concat([ratings1, ratings2, ratings3, ratings4, ratings5, ratings6,ratings7, ratings8], axis = 0)

ratings.shape
```

```python
#Generamos un columna date y eliminamos la columna timestamp, a su vez dejamos la columna date en el forma exigida por el trabajo

ratings['date'] = pd.to_datetime(ratings['timestamp'], unit = 's')
ratings['date'] = ratings['date'].dt.date
ratings = ratings.drop(columns='timestamp')
ratings
```

```python
#Renombramos la columna rating

ratings = ratings.rename(columns={'rating':'score'})
```

```python
#Lo exporto a csv

ratings.to_csv('../Datasets/score_movies.csv', index = False)
```

```python
#Calculo el promedio de cada pelicula 
score_prom = ratings.groupby("movieId")
score_prom = score_prom["score"].mean()
```

```python
#Lo pasamos a un dataframe y reseteamos el indice
df_score_prom = pd.DataFrame(score_prom)
df_score_prom.reset_index()
```
